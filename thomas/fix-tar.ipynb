```python
# Cellule 1 — inspecter les fichiers à la racine (ex: /mnt/data ou dossier courant)
import os

root = "/mnt/data"  # modifie si nécessaire, ou laisse "/mnt/data"
print("Exploration de :", root)
entries = sorted(os.listdir(root))
for e in entries:
    p = os.path.join(root, e)
    t = "DIR" if os.path.isdir(p) else "FILE"
    print(f"{t:4}  {e}")
    
# Afficher sous-arborescence succincte
print("\n--- Arborescence (2 niveaux) ---")
for e in entries:
    p = os.path.join(root, e)
    if os.path.isdir(p):
        sub = sorted(os.listdir(p))
        print(f"{e}/ : {len(sub)} éléments, exemples -> {sub[:5]}")
```
